```python
import pandas as pd
import numpy as np
```

# Industry RRI Weight

```python
weight = pd.read_csv("output/rriTSEWeight.csv")
weight.date = pd.to_datetime(weight.date)
weight = weight.set_index('date')
weight.columns = [str(col)+'_weight' for col in weight.columns]
weight = weight.asfreq('d').ffill()
weight = weight.fillna(0)
weight
```

# RRI Data

```python
data = pd.read_csv("output/rri.csv", index_col=0)
data.date = pd.to_datetime(data.date)
```

```python
rri = data.pivot_table(values='RRI', index='date', columns='company_id')
rri = rri.fillna(0)
rri.columns = [str(col)+'_rri' for col in rri.columns]
rri
```

# calculate company weighted RRI

```python
rri.head()
```

```python
weight.head()
```

```python
indRRI = pd.merge(rri, weight, on='date', how='left').ffill().fillna(0)
indRRI
```

# rri x weight

```python
res = dict()
for coid in [comp[:4] for comp in rri.columns]:
    df_comp = indRRI.filter(regex=coid).copy()

    if len(df_comp.columns) > 1:
        rri_scores = np.multiply(df_comp.iloc[:, 0], df_comp.iloc[:, 1])

    res[coid] = rri_scores.tolist()
```

```python
compIndRRI = pd.DataFrame(res)
compIndRRI
```

```python
# index 轉為日期
compIndRRI.index = pd.date_range(start=str(indRRI.index[0]), periods=len(compIndRRI), freq='d')
compIndRRI
```

# Industry RRI

```python
tse_data = pd.read_csv("data/TSE.csv")
tse_data.company_id = tse_data.company_id.apply(lambda x: str(x))
tseDict = tse_data.drop_duplicates(subset=['company_id'], keep='last')[['company_id', 'TSE']]
tseDict = tseDict.sort_values('company_id')
tseDict.dropna(inplace=True)
```

```python
rri_TSE = pd.DataFrame()
for ind in [i for i in tseDict.TSE.unique()]:
    lst = tseDict[tseDict['TSE']==ind]['company_id'].tolist()
    df_ind = compIndRRI[compIndRRI.columns.intersection(lst)]
    indrri = df_ind.sum(axis=1)
    rri_TSE[str(ind)] = indrri.tolist()

rri_TSE.index = pd.date_range(start=str(indRRI.index[0]), periods=len(compIndRRI), freq='d')
```
